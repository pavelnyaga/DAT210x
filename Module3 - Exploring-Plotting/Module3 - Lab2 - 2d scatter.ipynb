```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

Load up the wheat seeds dataset into a dataframe. We've stored a copy in the Datasets directory.

```python
df   = pd.read_csv('./Datasets/wheat.data', sep=',', index_col=0)
df.describe()
df.head(5)
```

Create a 2d scatter plot that graphs the `area` and `perimeter` features:

```python
# .. your fig = plt.figure()
fig = plt.figure()

ax = fig.add_subplot(211)
ax.set_xlabel('area')
ax.set_ylabel('perimeter')

ax.scatter(df.area, df.perimeter, c='r', marker='.')

ax2 = fig.add_subplot(212)
ax2.set_xlabel('groove')
ax2.set_ylabel('asymmetry')

ax2.scatter(df.groove, df.asymmetry, c='r', marker='.')
plt.show()
```

Create a 2d scatter plot that graphs the `groove` and `asymmetry` features:

```python
fig2 = plt.figure()

ax2 = fig2.add_subplot(111)
ax2.set_xlabel('groove')
ax2.set_ylabel('asymmetry')

ax2.scatter(df.groove, df.asymmetry, c='r', marker='.')
plt.show()
```

Create a 2d scatter plot that graphs the `compactness` and `width` features:

```python
fig3 = plt.figure()

ax3 = fig3.add_subplot(111)
ax3.set_xlabel('compactness')
ax3.set_ylabel('width')

ax3.scatter(df.compactness, df.width, c='b', marker='.')
plt.show()
```
